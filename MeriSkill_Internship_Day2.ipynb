```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Coffee Survey (Responses) - Form Responses 1.csv')
df.head()
```

```python
df.columns = df.columns.str.strip()

# prompt: just write the code where i can rename each column name of the dataset

df1 = df.rename(columns={
    'Timestamp': 'timestamp',
    'What is your age group?': 'age',
    'What is your occupation?': 'occupation',
    'What is your monthly household income?': 'household_income',
    'Where do you live in Mayur Vihar?': 'area',
    'How often do you consume coffee?': 'coffee_frequency',
    'Where do you usually buy your coffee?': 'place_to_buy',
    'How much do you typically spend on coffee per week?': 'spend_on_coffee_per_week',
    'What type of coffee do you prefer? (Select all that apply)': 'coffee_type',
    'How often do you buy coffee from a coffee shop?': 'coffee_shop_frequency',
    'What factors influence your choice of coffee shop? (Select all that apply)': 'influence_factors',
    'What time of day do you usually drink coffee?': 'time_to_drink',
    'How important is sustainability (e.g., eco-friendly packaging, fair trade coffee) when choosing a coffee brand?': 'sustainability',
    'Would you prefer a new coffee shop to offer any of the following? (Select all that apply)': 'new_shop_offers',
    'On a scale of 1 to 5, how satisfied are you with the current coffee options in Mayur Vihar?': 'satisfaction',
    'What improvements would you like to see in the coffee shops in Mayur Vihar?': 'improvements',
    'Would you visit a new coffee shop in Mayur Vihar if one opened?': 'new_shop_visit',
})

df1.head()
```

```python
age_counts = df1['age'].value_counts()


age_counts_df = age_counts.reset_index()
age_counts_df.columns = ['age', 'count']
total_responses = age_counts_df['count'].sum()


age_counts_df['percentage'] = (age_counts_df['count'] / total_responses) * 100

age_counts_df


import matplotlib.pyplot as plt

age = ['Under 18', '18-25','46-60', '26-35', 'Over 60']
percentage = [19.354839,54.838710,6.451613, 9.677419, 9.677419]

plt.pie(percentage, labels=age, autopct='%1.1f%%')
_ = plt.title('Percentage of Individuals in Each Age Group')
```

```python
income_mapping = {
    'Under ₹20,000': 10000,
    '₹20,000 - ₹50,000': 35000,
    '₹50,000 - ₹1,00,000': 75000,
    'Over ₹1,00,000': 125000
}

# Replace the household income strings with numeric values using the mapping
df1['household_income_numeric'] = df1['household_income'].map(income_mapping)
```

```python
mean_income_by_occupation = df1.groupby('occupation')['household_income_numeric'].mean()

mean_income_by_occupation
mean_income_by_occupation = df1.groupby('occupation')['household_income_numeric'].mean()

plt.figure(figsize=(8, 5))
mean_income_by_occupation.plot(kind='bar', color='skyblue')

plt.title('Mean Household Income by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Mean Household Income (₹)')
plt.xticks(rotation=45)


plt.show()
```

```python
frequency_counts = df1['coffee_frequency'].value_counts()
frequency_percentage = (frequency_counts / total_responses) * 100
frequency_percentage
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
ax = frequency_percentage.plot(kind='barh', color='darkblue')

plt.title('Frequency of Coffee Consumption', fontsize=16)
plt.xlabel('Percentage of Respondents (%)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

for index, value in enumerate(frequency_percentage):
    plt.text(value, index, f'{value:.2f}%', va='center')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df1 is already loaded
# Step 1: Ensure 'coffee_type' column is formatted correctly
df1['coffee_type'] = df1['coffee_type'].fillna('').astype(str)

# Step 2: Split the 'coffee_type' column into separate values
df1['coffee_type'] = df1['coffee_type'].str.split(',')

# Step 3: Expand the lists into separate rows using explode()
df1_exploded = df1.explode('coffee_type')

# Step 4: Strip any leading or trailing spaces from the split values
df1_exploded['coffee_type'] = df1_exploded['coffee_type'].apply(lambda x: x.strip())

# Step 5: Count the occurrences of each coffee type
coffee_count = df1_exploded['coffee_type'].value_counts()

# Step 6: Calculate the total number of responses
total_responses = len(df1_exploded)

# Step 7: Calculate the percentage for each coffee type
coff_percent = (coffee_count / total_responses) * 100

# Step 8: Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(coff_percent, labels=coff_percent.index, autopct='%1.2f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Percentage of Coffee Consumption Categories', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

spending_mapping = {
    'Under ₹100': 50,
    '₹100 - ₹300': 200,
    '₹300 - ₹500': 400,
    'Over ₹500': 600
}

df1['numeric_spending'] = df1['spend_on_coffee_per_week'].map(spending_mapping)

# Calculate the average weekly spending on coffee
average_weekly_spending = df1['numeric_spending'].mean()
print(f"Average weekly spending on coffee is: ₹{average_weekly_spending:.2f}")
spending_counts = df1['spend_on_coffee_per_week'].value_counts()
spending_counts
spending_percent = (spending_counts / total_responses) * 100
spending_percent
pd.crosstab(df1['age'], df1['spend_on_coffee_per_week']).plot(kind='bar', stacked=True)
```

```python
# Step 1: Split the 'influence_factors' column into separate values
df1['influence_factors'] = df1['influence_factors'].str.split(',')

# Step 2: Expand the lists into separate rows using explode()
df1_factors_exploded = df1.explode('influence_factors')

# Step 3: Strip any leading or trailing spaces from the split values
df1_factors_exploded['influence_factors'] = df1_factors_exploded['influence_factors'].str.strip()

# Step 4: Get the count of each unique factor
factors_count = df1_factors_exploded['influence_factors'].value_counts()

# Display the frequency count of each factor
factors_count
```

```python
plt.figure(figsize=(12, 8))
bar_plot = sns.barplot(y=factors_count.index, x=factors_count.values, palette='viridis')

# Add data labels on top of each bar
for index, value in enumerate(factors_count.values):
    bar_plot.text(value, index, f'{value}', color='black', ha='left', va='center')

plt.title('Influence Factors on Coffee Shop Choice')
plt.xlabel('Count of Respondents')
plt.ylabel('Influence Factors')
plt.show()
```

```python
dfq = df1['satisfaction'].value_counts()
dfq
```

```python


# Create a mapping of satisfaction levels
satisfaction_mapping = {
    '1 (Very dissatisfied)': 1,
    '2 (Dissatisfied)': 2,
    '3 (Neutral)': 3,
    '4 (Satisfied)': 4,
    '5 (Very satisfied)': 5
}

# Convert satisfaction descriptions to numeric scores
df1['satisfaction_score'] = df1['satisfaction'].map(satisfaction_mapping)

# Calculate the weighted sum of satisfaction scores
weighted_sum = (df1['satisfaction_score'] * df1['count']).sum()

# Calculate the total count of responses
total_count = df1['count'].sum()

# Calculate average satisfaction
average_satisfaction = weighted_sum / total_count

# Print the average satisfaction
print(f"Average Satisfaction: {average_satisfaction:.2f}")
plt.figure(figsize=(8, 6))
plt.bar(['Average Satisfaction'], [average_satisfaction], color='blue')

# Add data labels to the bar
plt.text(0, average_satisfaction, f'{average_satisfaction:.2f}', ha='center', va='bottom', fontsize=12)

plt.title('Average Satisfaction Score')
plt.ylabel('Satisfaction Score')
plt.ylim(0, 5)  # Set y-axis limit from 0 to 5 (as per satisfaction scale)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
